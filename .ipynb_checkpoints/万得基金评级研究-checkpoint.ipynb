```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', 50)
```

```python
df=pd.read_excel("fundrating.xls")
df.head(20)
```

```python
df_new=df.copy()
```

```python
df_rating=df_new.dropna().loc[:,'2019Q1':].corr()
df_rating
```

```python
corr=df_rating.loc['rating201812':,'2019Q1':'2021Q2']
mask = np.zeros_like(corr)
#i为行标，j为列标
#mask[np.triu_indices_from(mask)] = True
for i in range(0,10):
    for j in range(0,10):
        if((i-j<-3)|(i-j>0)):
            mask[i][j]=True
```

查看万得评级与基金连续4个季度收益率的相关系数

```python
#基金评级与区间收益率关系，行为基金评级的月份，列为基金评级下一个季度的区间收益率，我们观察评级后4个月的收益率
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(12, 8))
    ax =sns.heatmap(corr, mask=mask,annot=True,fmt=".2f",square=True)
```

```python
#将万得未覆盖评级的基金暂定为0星，用于作图
df_new.loc[:,'rating201812':]=df_new.loc[:,'rating201812':].fillna(0)
```

查看从2018年12月起，每次评级次季度收益率

```python
#查看从2018年12月起，每次评级次季度收益率
l=[]
for i in range(2,12):
    result=df_new.loc[:,df_new.columns[i]:].groupby(df_new.columns[i+10]).mean()[df_new.columns[i]]
    l.append(result.values)
pd.DataFrame(data=l,index=df_new.columns[12:])
```
